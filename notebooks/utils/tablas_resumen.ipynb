```python
import glob
import pandas as pd

path = r'C:\Users\guill\Documents\1Repos\algoritmos-nlp\notebooks\resultados\notas\*.csv'
files = glob.glob(path)

files
```

```python
df_list = []
for file in files:
    df = pd.read_csv(file, sep=",", engine="c", on_bad_lines='skip')
    df_list.append(df)

all_data = pd.concat(df_list, axis=0, ignore_index=True)
all_data
```

```python
from datetime import timedelta



# Convertir la columna Time en formato h.m.s
all_data['runtime_hms'] = all_data['tiempo_seg'].apply(lambda x: str(timedelta(seconds=x)))
```
